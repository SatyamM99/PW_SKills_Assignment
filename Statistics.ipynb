1. What is the difference between descriptive statistics and inferential
statistics? Explain with examples.

The difference lies in their purpose and scope of analysis: Descriptive statistics summarize data, while inferential statistics draw conclusions about a larger population.


**Descriptive Statistics**

Descriptive statistics summarizes, organizes, and simplifies characteristics of a data set (a sample or an entire population). It describes what is in the data.

Goal: To present data in a meaningful way, often through graphs or summaries.

Scope: Limited to the actual data collected. You cannot use it to draw conclusions beyond that data.

Examples:

Measures of Central Tendency: Calculating the average (mean) house price in your neighborhood.

Measures of Variability: Reporting the range or standard deviation of student test scores in a single class.


**Inferential Statistics**

Inferential statistics uses data from a sample to make generalizations, predictions, or inferences about a much larger population. It helps you draw conclusions beyond the immediate data.

Goal: To test hypotheses, estimate population parameters, and determine if results are statistically significant.

Scope: Extends from the sample to the entire population.

Examples:

Hypothesis Testing: Using the average test scores from one class (sample) to determine if a new teaching method is effective for all students (population).

Estimation: Using a political poll (sample) to predict the final election results for the entire country (population), providing a margin of error.

Sampling in statistics is the process of selecting a subset of individuals or observations from a larger population to estimate characteristics of that entire population. The goal is to choose a sample that is representative of the population to ensure the results are generalizable.

3.  Define mean, median, and mode. Explain why these measures of central
tendency are important.

Mean, median, and mode are the three main measures of central tendency, used to find the typical or central value in a dataset.

**Mean (Average)**- The arithmetic average of all values.	Sum of all values divided by the total count of values.

**Median (Middle)**-	The middle value in a dataset when the values are arranged in order.	The value that separates the upper half from the lower half of the data.

**Mode (Most Frequent)**-	The value that appears most frequently in a dataset.	The value with the highest frequency.


These measures are crucial because they provide a single, representative value that summarizes an entire distribution, which is useful for:

Summarizing Data: They offer a quick, simplified understanding of where the data clusters (e.g., the average salary, the typical household size).

Comparison: They allow for easy comparison between different groups or time periods (e.g., comparing the mean test score of two different classes).

Detecting Skewness: Comparing the mean, median, and mode helps identify the shape of the distribution and detect skewness (asymmetry).

If Mean > Median > Mode, the distribution is right-skewed.

If Mean ≈ Median ≈ Mode, the distribution is roughly symmetrical (normal).

Handling Outliers: The median is particularly important because it is robust to outliers (extreme values), unlike the mean, which can be heavily pulled by them.


4. Explain skewness and kurtosis. What does a positive skew imply about the
data?

Skewness and kurtosis are two measures used to describe the shape of a probability distribution.

**Skewness**

Skewness measures the asymmetry of a distribution. It indicates whether the data is concentrated on one side and how far the tail extends on the other side relative to the mean.

Zero Skew: The distribution is perfectly symmetrical (like a normal distribution), and the Mean ≈ Median ≈ Mode.

Positive Skew (Right Skew): The long tail extends to the right.

Negative Skew (Left Skew): The long tail extends to the left.

**Kurtosis**

Kurtosis measures the "tailedness" of a distribution, indicating how heavily the tails are weighted compared to a normal distribution. It describes the peak's sharpness and the flatness of the tails.

Mesokurtic: Similar to a normal distribution, with a kurtosis value of approximately zero (Excess Kurtosis = 0).

Leptokurtic: Has fat, heavy tails and a high, sharp peak. It indicates more outliers and a higher probability of extreme values (Excess Kurtosis > 0).

Platykurtic: Has thin, light tails and a flatter peak. It indicates fewer outliers than a normal distribution (Excess Kurtosis < 0).

5. Implement a Python program to compute the mean, median, and mode of
a given list of numbers.

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


```python
import statistics

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Compute Mean
mean_value = statistics.mean(numbers)

# Compute Median
median_value = statistics.median(numbers)

# Compute Mode
mode_value = statistics.mode(numbers)

print(f"List of numbers: {numbers}")
print("-" * 35)
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Mode: {mode_value}")
```

6. Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]


```python
import numpy as np

# Given lists
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert to numpy arrays for calculation
X = np.array(list_x)
Y = np.array(list_y)

# Compute Covariance
cov_matrix = np.cov(X, Y)
covariance = cov_matrix[0, 1]

# Compute Correlation Coefficient
corr_matrix = np.corrcoef(X, Y)
correlation_coefficient = corr_matrix[0, 1]

print("--- Statistical Measures ---")
print(f"Covariance (Cov(X, Y)): {covariance:.2f}")
print(f"Correlation Coefficient (r): {correlation_coefficient:.4f}")
```

This result indicates a very strong positive linear relationship between X and Y, as the correlation coefficient (r) is very close to +1.

7. Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Given list of numbers
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Calculate Quartiles and IQR for outlier identification
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR
outliers = [x for x in data if x < lower_fence or x > upper_fence]

# Create Boxplot
plt.figure(figsize=(6, 4))
plt.boxplot(data, vert=False, patch_artist=True)
plt.title('Boxplot to Identify Outliers')
plt.yticks([1], ['Data'])
plt.xlabel('Values')
plt.grid(axis='x', linestyle=':')
plt.savefig('boxplot_outliers.png')

print("--- Outlier Detection (1.5 * IQR Rule) ---")
print(f"Q1 (25th percentile): {Q1:.2f}")
print(f"Q3 (75th percentile): {Q3:.2f}")
print(f"IQR (Q3 - Q1): {IQR:.2f}")
print(f"Lower Fence: {lower_fence:.2f}")
print(f"Upper Fence: {upper_fence:.2f}")
print(f"\nIdentified Outliers: {outliers}")
```

The outlier is 35.

Boxplot Meaning: The central box spans the Interquartile Range (IQR, from Q1=17.25 to Q3=23.25), representing the middle 50% of the data. The line inside is the median.

Outlier Rule: Values lying outside the calculated fences (Lower Fence: 8.25, Upper Fence: 32.25) are considered outliers by the standard 1.5×IQR rule.

Conclusion: The value 35 is the only data point greater than the Upper Fence of 32.25, thus it is identified as an outlier.

8. You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and daily sales.

Explain how you would use covariance and correlation to explore this
relationship.

Write Python code to compute the correlation between the two lists:

advertising_spend = [200, 250, 300, 400, 500]

daily_sales = [2200, 2450, 2750, 3200, 4000]


```python
import numpy as np

# Data provided
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert lists to NumPy arrays
X = np.array(advertising_spend)
Y = np.array(daily_sales)

# Compute the correlation matrix
# The correlation coefficient (r) is found at index [0, 1] or [1, 0]
corr_matrix = np.corrcoef(X, Y)
correlation_coefficient = corr_matrix[0, 1]

print("--- Correlation Analysis ---")
print(f"Advertising Spend (X): {advertising_spend}")
print(f"Daily Sales (Y): {daily_sales}")
print("-" * 30)
print(f"Correlation Coefficient (r): {correlation_coefficient:.4f}")
```

*Covariance*: Measures the direction of the linear relationship between two variables.

Interpretation: A positive covariance means that as advertising spend increases, daily sales also tend to increase (a positive relationship). A negative value means they move in opposite directions.

Limitation: The value itself is unbounded, making it hard to judge the strength of the relationship (e.g., is 1000 a strong covariance?).


**Correlation Coefficient (r)**:Measures both the direction and strength of the linear relationship.

Interpretation: The value is standardized to be between −1.0 and +1.0.

r≈+1.0: Very strong positive relationship (high spend leads to high sales).

r≈0: No linear relationship.

Conclusion for Marketing: The correlation coefficient provides the marketing team with a clear, standardized metric to determine how reliably advertising spend predicts sales.

9. Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.

Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you would use.

Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

```python
import matplotlib.pyplot as plt
import numpy as np

# Survey data (scores 1-10)
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# 1. Create the histogram
plt.figure(figsize=(7, 5))
# Use fixed bins corresponding to the integer scores 1 to 10
# bins = np.arange(3.5, 11.5, 1.0) ensures centered bars for each integer
bins = np.arange(min(survey_scores) - 0.5, max(survey_scores) + 1.5, 1)

plt.hist(survey_scores, bins=bins, color='skyblue', edgecolor='black', rwidth=0.8)

# 2. Add labels and title
plt.title('Distribution of Customer Satisfaction Scores (1-10)')
plt.xlabel('Satisfaction Score')
plt.ylabel('Frequency')
plt.xticks(range(4, 11)) # Ensure x-axis shows integer scores clearly
plt.grid(axis='y', alpha=0.5)

# 3. Save the plot
plt.savefig('satisfaction_scores_histogram.png')
print("Histogram saved as satisfaction_scores_histogram.png")
```

**Mean & Median**-	To find the central tendency (average and middle score). A high mean suggests overall satisfaction is good.

**Standard Deviation**-	To measure the spread (variability) of the scores. A low value means customers generally agree on the score; a high value means scores are polarized.

**Histogram**-	To visualize the entire frequency distribution. This shows the shape, central peak, and any clustering of scores (e.g., if most scores are 7-8, or if there's a bimodal distribution at 5 and 10).
