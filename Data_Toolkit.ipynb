1 What is NumPy, and why is it widely used in Python?

NumPy is a fundamental Python library for numerical computing. It's widely used because it provides a powerful ndarray object for storing and manipulating large, multidimensional arrays and matrices efficiently.

2 How does broadcasting work in NumPy?

Broadcasting is a mechanism that allows NumPy to perform operations on arrays of different shapes. It automatically stretches the smaller array to match the shape of the larger array for the operation, without actually duplicating the data in memory.

3 What is a Pandas DataFrame?

A Pandas DataFrame is a 2-dimensional, size-mutable, and potentially heterogeneous tabular data structure. It's similar to a spreadsheet or a SQL table, with labeled axes (rows and columns).

4 Explain the use of the groupby() method in Pandas.

The groupby() method is used to split the data into groups based on some criteria. You can then apply a function (like sum(), mean(), count(), etc.) to each group independently.

5 Why is Seaborn preferred for statistical visualizations?

Seaborn is a data visualization library based on Matplotlib. It's preferred for statistical visualizations because it comes with a high-level interface for drawing attractive and informative statistical graphics. It handles many of the complexities of data plotting automatically.

6 What are the differences between NumPy arrays and Python lists?

NumPy arrays are fixed-size, homogeneous (all elements of the same type), and support vectorized operations. Python lists are dynamic, heterogeneous, and do not support vectorized operations, which makes NumPy arrays much faster for numerical computations.

7 What is a heatmap, and when should it be used?

A heatmap is a graphical representation of data where values are represented by color. It's used to visualize data in a matrix format and is excellent for showing the magnitude of a phenomenon in two dimensions, such as a correlation matrix.

8 What does the term "vectorized operation" mean in NumPy?

A vectorized operation is an operation that is applied to all elements of an array at once, without the need for an explicit for loop. This is much faster and more concise than iterating through elements one by one.

9 How does Matplotlib differ from Plotly?

Matplotlib is a low-level plotting library that gives you fine-grained control over every aspect of a plot. Plotly is a high-level, interactive visualization library that creates web-based plots that can be zoomed, panned, and hovered over.

10 What is the significance of hierarchical indexing in Pandas?

Hierarchical indexing, or MultiIndex, allows you to have multiple index levels on an axis. This is significant because it lets you handle and represent higher-dimensional data in a two-dimensional DataFrame, making complex data easier to manage.

11 What is the role of Seaborn's pairplot() function?

The pairplot() function creates a grid of plots, showing pairwise relationships between variables in a dataset. It's a quick and effective way to visualize the distribution of each variable and the correlation between them.

12 What is the purpose of the describe() function in Pandas?

The describe() function generates descriptive statistics of a DataFrame's numerical columns, including count, mean, standard deviation, min, max, and quartiles. It's a great tool for quickly summarizing data.

13 Why is handling missing data important in Pandas?

Handling missing data is important because it can skew statistical analysis and lead to inaccurate conclusions. Pandas provides several ways to deal with missing data, such as removing rows or columns, or filling the missing values with a mean, median, or other value.

15 What are the benefits of using Plotly for data visualization?

Plotly's main benefits are its interactivity, which allows users to explore data dynamically, and its ability to create professional-quality, web-based visualizations that can be easily shared.

16 How does NumPy handle multidimensional arrays?

NumPy's ndarray object is specifically designed to handle multidimensional arrays efficiently. You can create arrays with any number of dimensions, and NumPy provides intuitive indexing and slicing to access elements.

17 What is the role of Bokeh in data visualization?

Bokeh is a library that creates interactive, web-based plots and dashboards. Its primary role is to enable the creation of highly-customizable and scalable visualizations that can be embedded in web applications.

18 Explain the difference between apply() and map() in Pandas.

The apply() method is more general and can work on an entire DataFrame, a row, or a column. The map() method is a Series-only method that is used to substitute each value in a Series with another value.

19 What are some advanced features of NumPy?

Advanced features of NumPy include broadcasting, advanced indexing (like boolean indexing), linear algebra operations, and Fourier transforms.

20 How does Pandas simplify time series analysis?

Pandas simplifies time series analysis by providing specialized tools and data structures, such as the DatetimeIndex, which allows for easy resampling, shifting, and a wide range of time-based operations.

21 What is the role of a pivot table in Pandas?

A pivot table is used to summarize and rearrange data in a DataFrame. It aggregates data based on one or more columns, providing a new view of the data that's useful for analysis.

22 Why is NumPy's array slicing faster than Python's list slicing?

NumPy's array slicing is faster because it creates a "view" of the original array, meaning it doesn't copy the data. Python's list slicing, on the other hand, creates a new list, which is a more memory-intensive and slower operation.

23 What are some common use cases for Seaborn?

Common use cases for Seaborn include creating visualizations for a single variable's distribution, exploring relationships between multiple variables with scatter plots, and visualizing categorical data. It is often used for initial data exploration and statistical modeling.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# 1. 2D array & sum of rows
arr = np.array([[1, 2], [3, 4]])
print(arr.sum(axis=1))
```

```python
# 2. Mean of column
df = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
print(df["A"].mean())
```

```python
# 3. Scatter plot
plt.scatter([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# 4. Correlation heatmap
data = np.random.rand(5, 5)
sns.heatmap(pd.DataFrame(data).corr(), annot=True)
plt.show()
```

```python
# 5. Bar plot Plotly
fig = px.bar(x=["A", "B"], y=[1, 2])
fig.show()
```

```python
# 6. Add column
df["C"] = df["A"] * 2
print(df)
```

```python
# 7. Element-wise multiplication
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a * b)
```

```python
# 8. Multi-line plot
plt.plot([1, 2], [3, 4], label="Line 1")
plt.plot([1, 2], [4, 6], label="Line 2")
plt.legend()
plt.show()
```

```python
# 9. Filter DataFrame
print(df[df["A"] > 1])
```

```python
# 10. Histogram
sns.histplot([1, 2, 2, 3])
plt.show()
```

```python
# 11. Matrix multiplication
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5, 6], [7, 8]])
print(np.dot(m1, m2))
```
