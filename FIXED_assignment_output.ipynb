1. What is a Vector Database (VectorDB) and how is it different from
traditional databases?

Ans- A Vector Database is a specialized storage system designed to manage and index high-dimensional data represented as vectors (mathematical arrays of numbers). These vectors, often called embeddings, represent the "meaning" or features of unstructured data like text, images, or audio.

Key Differences

**Data Representation**: Traditional databases store data in tables (rows/columns) or documents. VectorDBs store data as numerical embeddings generated by machine learning models.

**Search Method**: Traditional systems use keyword matching or exact values. VectorDBs use similarity search (e.g., Cosine Similarity) to find data points that are mathematically "close" to each other in high-dimensional space.

**Data Type**: Traditional databases are optimized for structured data (strings, integers). VectorDBs are built for unstructured data (finding a similar image or a semantically related sentence).

**Complexity**: We use traditional databases for discrete, logical queries (e.g., "Find all orders over $50"). We use VectorDBs for fuzzy, contextual queries (e.g., "Find songs that sound like this one").

2.Explain the various types of VectorDBs available and describe their
suitability for different use cases.

Ans- Modern vector databases are generally categorized into three types based on their architecture and how they integrate with existing systems.

1.**Specialized (Standalone) Vector Databases**

These are built from the ground up specifically to handle high-dimensional vector data and similarity searches.

Examples: Pinecone, Milvus, Weaviate, Qdrant, Chroma.

Suitability:

**High-Scale Production**: Best when managing millions or billions of vectors requiring sub-100ms latency.

**AI-Native Apps**: Ideal for complex RAG (Retrieval-Augmented Generation) pipelines and autonomous AI agents.

Advanced Filtering: Useful when you need to combine semantic search with complex metadata filtering (e.g., "Find similar images uploaded in the last 24 hours").

2.**Vector Libraries**

These are not full databases but rather specialized software libraries that implement similarity search algorithms (like HNSW or IVF).

Examples: FAISS (Meta), Annoy (Spotify), ScaNN (Google).

Suitability:

**In-Memory Tasks**: Best for small to medium datasets that can fit entirely in a server's RAM.

**Prototyping**: Ideal for researchers or developers building a custom search service where features like "backups" or "multi-tenancy" aren't required.

Embedded Use: When the search capability needs to be baked directly into a specific application rather than living on a separate server.

3.**Vector-Enabled Traditional Databases (Extensions)**

Existing relational or NoSQL databases that have added vector support as a feature or extension.

Examples: pgvector (PostgreSQL), Elasticsearch, MongoDB Atlas Vector Search, Redis.

Suitability:

**Unified Infrastructure**: Best for teams that want to keep their structured data and vectors in the same place to avoid managing extra tools.

**Hybrid Queries**: Ideal when your application heavily relies on traditional relational data (SQL) alongside semantic search.

**Moderate Scale**: Perfect for applications with under 10–20 million vectors where the overhead of a specialized system isn't yet justified.

text generation task.

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "gpt2" # Example: a lightweight model for demonstration
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Ensure the tokenizer has a padding token
tokenizer.pad_token = tokenizer.eos_token
```

```python
from datasets import load_dataset

# Load a sample dataset (e.g., wiki_text or a custom JSON)
dataset = load_dataset("wikitext", "wikitext-2-raw-v1", split="train[:1%]")

def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, padding="max_length", max_length=128)

tokenized_datasets = dataset.map(tokenize_function, batched=True, remove_columns=["text"])
```

```python
from transformers import TrainingArguments, Trainer, DataCollatorForLanguageModeling

# Data collator specifically for Causal Language Modeling
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=4,
    num_train_epochs=1,
    save_steps=100,
    logging_steps=10,
    push_to_hub=False, # Set to True to share on Hugging Face Hub
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets,
    data_collator=data_collator,
)

trainer.train()
```

```python
# 1. Save the final model and tokenizer to a dedicated folder
final_model_path = "./final_text_gen_model"
trainer.save_model(final_model_path)
tokenizer.save_pretrained(final_model_path)

# 2. Load the objects using the Auto classes (more reliable than pipeline paths)
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model = AutoModelForCausalLM.from_pretrained(final_model_path)
tokenizer = AutoTokenizer.from_pretrained(final_model_path)

# 3. Initialize pipeline with the objects directly
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer
)

# 4. Test generation
print(generator("The future of AI is", max_length=50, truncation=True))
```

```python
import ollama
import chromadb
from chromadb.utils import embedding_functions

# 1. Initialize an Ephemeral (In-memory) Client
client = chromadb.Client()

# 2. Setup Embedding Function (runs locally in Colab)
ef = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-MiniLM-L6-v2")
collection = client.get_or_create_collection(name="colab_docs", embedding_function=ef)

# 3. Add Sample Documents
documents = [
    "The 2026 World Cup will be hosted by Canada, Mexico, and the United States.",
    "Llama 3 is a state-of-the-art open-source LLM developed by Meta.",
    "RAG works by retrieving relevant documents before generating an answer."
]
collection.add(documents=documents, ids=["id1", "id2", "id3"])

# 4. Define the Retrieval and Generation Logic
def run_rag(user_query):
    # Search for the most relevant context
    results = collection.query(query_texts=[user_query], n_results=1)
    context = results["documents"][0][0]

    # Combine context with the query for the LLM
    prompt = f"Context: {context}\n\nQuestion: {user_query}\nAnswer:"

    response = ollama.generate(model="llama3:8b", prompt=prompt)
    return response['response'], context

# 5. Execution
query = "Who made Llama 3?"
answer, source = run_rag(query)

print(f"Source Context: {source}")
print(f"Llama 3 Answer: {answer}")
```

10. A health-tech startup wants to build a chatbot that can answer user
queries based on medical research articles. Propose and explain a solution using
Hugging Face models for understanding, VectorDB for retrieval, and Ollama for
generation.

To build a medical research chatbot, we implement a RAG (Retrieval-Augmented Generation) pipeline that combines domain-specific embeddings with a local LLM for privacy and accuracy.

**Proposed Architecture**

**Understanding (Hugging Face)**: We use a specialized medical embedding model (like BAAI/bge-small-en-v1.5 or a PubMed-trained model) to convert complex research text into high-dimensional vectors.

**Retrieval (VectorDB)**: ChromaDB stores these embeddings. When a user asks a question, we perform a similarity search to find the most relevant snippets from the research articles.

**Generation (Ollama):** We pass the retrieved snippets as context to Llama3 via Ollama. This ensures the chatbot answers based on the provided research rather than general knowledge.

```python
# Install system dependencies
!sudo apt-get install -y zstd
!curl -fsSL https://ollama.com/install.sh | sh

# Start Ollama in background and pull model
import subprocess, time
subprocess.Popen(['ollama', 'serve'])
time.sleep(10)
!ollama pull llama3:8b

# Install Python libraries
!pip install ollama chromadb sentence-transformers
```

```python
import ollama
import chromadb
from chromadb.utils import embedding_functions

# 1. Initialize VectorDB with a Hugging Face Medical-friendly model
ef = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-MiniLM-L6-v2")
client = chromadb.Client()
collection = client.get_or_create_collection(name="medical_research", embedding_function=ef)

# 2. Add Sample Medical Research Snippets
research_articles = [
    "Study A: Ibuprofen is effective for minor pain but can cause gastric upset in 10% of patients.",
    "Study B: Recent trials show Vitamin D3 supplementation improves bone density in elderly subjects.",
    "Study C: Telemedicine increased healthcare access by 30% in rural areas during 2023."
]
collection.add(documents=research_articles, ids=["doc1", "doc2", "doc3"])

# 3. Query & Generate Function
def medical_chatbot(user_query):
    # Retrieve top 2 relevant snippets
    results = collection.query(query_texts=[user_query], n_results=2)
    context = " ".join(results["documents"][0])

    # Generate response with Ollama
    prompt = f"System: Answer using only the provided research.\nContext: {context}\nQuestion: {user_query}"
    response = ollama.generate(model="llama3:8b", prompt=prompt)

    return response['response']

# 4. Test
print(medical_chatbot("What are the side effects of Ibuprofen?"))
```
