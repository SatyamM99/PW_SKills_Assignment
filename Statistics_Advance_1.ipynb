1. What is a random variable in probability theory?

A random variable in probability theory is a function that assigns a numerical value to each possible outcome of a random phenomenon or experiment.

It is a variable whose possible values are determined by chance.

2. What are the types of random variables?

The two main types of random variables are:

**Discrete Random Variables:**

Variables whose possible values are countable (finite or countably infinite).

They typically result from counting.

Example: The number of defective items in a batch, the number of heads in four coin flips.

**Continuous Random Variables:**

Variables that can take any value within a given range or interval.

They typically result from measuring.

Example: The height of a person, the temperature of a room, the time it takes for a customer to complete a transaction.

3. Explain the difference between discrete and continuous distributions.

**Discrete Distributions**

Variable Type: Describes Discrete Random Variables (countable values).

Key Function: Probability Mass Function (PMF): Gives the probability that the variable exactly equals a specific value.

Probability: Probability is calculated for a single, specific value (P(X=x)).

Example: Binomial, Poisson

**Continuous Distributions**

Variable Type: Describes Continuous Random Variables (measurable values over a range).

Key Function: Probability Density Function (PDF): Gives the relative likelihood of the variable taking on a specific value.

Probability: Probability is calculated over an interval or range (P(a<X<b)), as the probability of any single, exact point is zero.

Example: Normal (Gaussian), Uniform, Exponential.

4.  What is a binomial distribution, and how is it used in probability?

A Binomial Distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes (success or failure).

Key Uses:

**Modeling Binary Outcomes:** It is used to analyze scenarios where an experiment is repeated multiple times, and the outcome of interest is binary (e.g., pass/fail, yes/no, coin flip).

**Calculating Probability:** It helps calculate the probability of getting exactly k successes in n trials, given the probability of success, p, on any single trial.

5. What is the standard normal distribution, and why is it important?

The Standard Normal Distribution is a special case of the Normal (Gaussian) distribution where the mean (μ) is 0 and the standard deviation (σ) is 1.

Importance:

Standardization (Z-Scores): Any normal distribution can be converted into the standard normal distribution using the Z-score formula. This process, called standardization, allows data from any normal distribution to be easily compared.


Probability Lookups: It allows statisticians to use a single Z-table (or Z-score calculator) to find the probability (area under the curve) associated with any value, regardless of the original mean and standard deviation.

Foundation: It serves as the basis for many critical statistical tests (like t-tests, ANOVA) and the Central Limit Theorem.

6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

The Central Limit Theorem (CLT) states that if you take large enough random samples from any population, the distribution of the sample means will tend to become a Normal Distribution, regardless of the shape of the original population distribution.

Importance:

Enables Inference: It allows us to use the powerful, well-understood properties of the Normal Distribution for inferential statistics (like constructing confidence intervals and performing hypothesis tests) even when the original population data is non-normal (skewed or uniform).

Foundation of Statistics: It justifies using statistical methods that rely on normality, such as the z-test and t-test, especially when dealing with means derived from large samples (n≥30 is a common heuristic).

Predictability: It explains why many natural phenomena and random processes (like measurement errors or test scores) often follow a normal distribution.

7. What is the significance of confidence intervals in statistical analysis?

Confidence intervals (CIs) are crucial because they provide a range of values that is likely to contain the true value of a population parameter (like the mean).

Quantifies Uncertainty: They quantify the uncertainty or precision of a sample estimate.

Reliability: A 95% confidence interval, for instance, means that if you were to repeat the sampling process many times, 95% of the resulting intervals would contain the true population parameter.

Better than Point Estimates: They are superior to a single point estimate (like a sample mean) because they give a clear, defined boundary for the possible error.

8. What is the concept of expected value in a probability distribution?

The Expected Value (E|X) of a probability distribution is the long-run average value of a random variable.

Concept: It represents the mean (μ) of the distribution.

Calculation: It's calculated by multiplying each possible outcome by its probability and summing the results.

Significance: It tells you what outcome you can expect to observe on average if you repeat the random process many times.

9. Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standarddeviation using NumPy, and draw a histogram to visualize the distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Normal Distribution
mean = 50
std_dev = 5
num_samples = 1000

# 1. Generate random numbers from a Normal Distribution
np.random.seed(42) # for reproducibility
data = np.random.normal(loc=mean, scale=std_dev, size=num_samples)

# 2. Compute the mean and standard deviation of the generated data
computed_mean = np.mean(data)
computed_std = np.std(data)

# 3. Draw a histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, edgecolor='black', alpha=0.7, color='skyblue', density=True)
plt.axvline(computed_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean ({computed_mean:.2f})')
plt.title(f'Histogram of {num_samples} Samples from N(μ={mean}, σ={std_dev})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(axis='y', alpha=0.5)
plt.savefig('normal_distribution_histogram.png')

# 4. Print results
print("--- Generated Data Analysis ---")
print(f"Theoretical Mean (μ): {mean}")
print(f"Theoretical Std Dev (σ): {std_dev}")
print("-" * 35)
print(f"Computed Sample Mean: {computed_mean:.4f}")
print(f"Computed Sample Std Dev: {computed_std:.4f}")
```

10. You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.

Write the Python code to compute the mean sales and its confidence interval.

The Central Limit Theorem (CLT) is applied because:

**Unknown Population Distribution**: We assume the population sales data is unknown or non-normal.

**Sampling Distribution is Normal**: The CLT guarantees that the distribution of sample means (the sampling distribution) will be approximately normal.

**Inference**: This normality allows us to use the t-distribution (since the population standard deviation is unknown) to calculate the 95% confidence interval, reliably estimating the true average daily sales for the entire two-year period.

```python
import numpy as np
from scipy import stats

# Daily sales data for 2 years (sample data provided)
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Confidence level
confidence_level = 0.95

# 1. Compute Sample Statistics
sample_mean = np.mean(daily_sales)
sample_std = np.std(daily_sales, ddof=1) # Use ddof=1 for sample standard deviation
sample_size = len(daily_sales)

# 2. Compute the Standard Error of the Mean (SEM)
sem = sample_std / np.sqrt(sample_size)

# 3. Calculate the Confidence Interval using the t-distribution
# Degrees of freedom (df)
df = sample_size - 1

# Calculate the critical t-value for a 95% confidence level
# ppf is the Percent Point Function (inverse of CDF)
t_critical = stats.t.ppf(1 - (1 - confidence_level) / 2, df)

# Calculate the margin of error (ME)
margin_of_error = t_critical * sem

# Calculate the confidence interval bounds
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print("--- Sales Confidence Interval Analysis ---")
print(f"Sample Mean Sales: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Sample Size (n): {sample_size}")
print(f"T-Critical Value (95% CI, df={df}): {t_critical:.4f}")
print("-" * 40)
print(f"95% Confidence Interval for Mean Sales:")
print(f"({ci_lower:.2f}, {ci_upper:.2f})")
```
