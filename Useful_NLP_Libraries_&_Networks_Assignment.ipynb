1. Compare and contrast NLTK and spaCy in terms of features, ease of use,
and performance.

Ans- **Features & Philosophy**

NLTK: A research-focused library providing a wide collection of algorithms for experiment and teaching.

spaCy: A production-focused library providing "opinionated" (one best) state-of-the-art models for industrial use.

**Ease of Use**

NLTK: String-based and modular. You must manually call and chain every step (e.g., download data, then tokenize, then tag).

spaCy: Object-oriented and streamlined. A single call processes text through an entire pipeline (Tokenization $\rightarrow$ POS $\rightarrow$ NER).

**Performance**

NLTK: Slower; written in pure Python.

spaCy: Much faster; core components are written in Cython for high-speed processing of large datasets.

2. What is TextBlob and how does it simplify common NLP tasks like
sentiment analysis and translation?

Ans- TextBlob is a beginner-friendly Python library that provides a simple API for common NLP tasks. It acts as a "wrapper," making complex tools like NLTK and Pattern much easier to use.

**How It Simplifies Tasks**

Sentiment Analysis: You can get the emotional tone of a sentence with one command. It returns Polarity (how positive/negative, from -1 to 1) and Subjectivity (how much is opinion vs. fact).

Translation: It provides built-in methods for language detection and translation (powered by Google Translate) without requiring you to train or load heavy models.

Spelling Correction: It includes a simple .correct() method to automatically fix typos in a string.

Intuitive Syntax: Unlike other libraries, you treat text as a Python object. Instead of complex functions, you simply access attributes like .words, .tags, or .noun_phrases.

4. Describe the architecture and functioning of a Recurrent Natural Network
(RNN).


Ans- An RNN is a neural network designed for sequential data (like sentences or time series). Unlike standard networks, it has "memory" because its output depends on both the current input and what it learned from previous steps.

**Core Architecture**

The Hidden State (h): Acts as the network's internal memory. It stores information from previous time steps.

Recurrent Loops: The network passes the hidden state back into itself at each step.Shared Weights: The same mathematical operations are applied to every element in the sequence, allowing it to process sequences of any length.

**How it Functions**

Input: The network takes the current word/data point ($x_t$) and the previous memory ($h_{t-1}$).

Processing: It combines them using a mathematical formula to create a new hidden state ($h_t$).

Output: It produces a result ($y_t$) based on that updated memory.Repeat: This continues until the entire sequence is processed.

**Main Limitation**

Short-term Memory: RNNs struggle to remember information from a long time ago in a sequence (the Vanishing Gradient problem). This is why they are often replaced by LSTMs or Transformers in modern AI.

5. What is the key difference between LSTM and GRU networks in NLP
applications?


Ans- The key difference is structural complexity. GRUs are a simplified version of LSTMs, designed to be faster while maintaining similar performance.

**Gating Mechanism:**

LSTM: Uses three gates (Input, Forget, and Output).

GRU: Uses two gates (Reset and Update). It combines the input and forget gates into a single "Update" gate.

**Internal State:**

LSTM: Maintains a separate Cell State ($c_t$) for long-term memory and a Hidden State ($h_t$) for short-term memory.

GRU: Merges these into a single Hidden State.

**Performance & Speed:**

LSTM: More parameters; slower to train but better for very long, complex sequences.

GRU: Fewer parameters; faster to train and computationally more efficient.

6.  Write a Python program using TextBlob to perform sentiment analysis on
the following paragraph of text:

“I had a great experience using the new mobile banking app. The interface is intuitive, and customer support was quick to resolve my issue. However, the app did crash once during a transaction, which was frustrating"

Your program should print out the polarity and subjectivity scores

```python
from textblob import TextBlob
import nltk

# Download required data for TextBlob
nltk.download('punkt_tab')

text = """I had a great experience using the new mobile banking app. The interface is intuitive,
and customer support was quick to resolve my issue. However, the app did crash once
during a transaction, which was frustrating."""

# Create a TextBlob object
blob = TextBlob(text)

# Perform sentiment analysis
polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity

print(f"Polarity Score: {polarity:.2f}")      # -1.0 (Negative) to 1.0 (Positive)
print(f"Subjectivity Score: {subjectivity:.2f}") # 0.0 (Objective) to 1.0 (Subjective)
```

7. Given the sample paragraph below, perform string tokenization and
frequency distribution using Python and NLTK:

“Natural Language Processing (NLP) is a fascinating field that combines linguistics, computer science, and artificial intelligence. It enables machines to understand, interpret, and generate human language. Applications of NLP include chatbots, sentiment analysis, and machine translation. As technology advances, the role of NLP in modern solutions is becoming increasingly critical.”

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Download required resource
nltk.download('punkt_tab')

text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical."""

# 1. Tokenization (Lowercased for accurate frequency counting)
tokens = word_tokenize(text.lower())

# 2. Frequency Distribution
fdist = FreqDist(tokens)

print(f"Total Tokens: {len(tokens)}")
print("-" * 30)
print("Top 5 Frequencies:", fdist.most_common(5))
```

8. Implement a basic LSTM model in Keras for a text classification task using
the following dummy dataset. Your model should classify sentences as either positive (1) or negative (0).

Dataset

texts = [

“I love this project”, #Positive

“This is an amazing experience”, #Positive

“I hate waiting in line”, #Negative

“This is the worst service”, #Negative

“Absolutely fantastic!” #Positive

]

labels = [1, 1, 0, 0, 1]

Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on this data. You may use Keras with TensorFlow backend.

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# 1. Dataset
texts = ["I love this project", "This is an amazing experience",
         "I hate waiting in line", "This is the worst service", "Absolutely fantastic!"]
labels = np.array([1, 1, 0, 0, 1])

# 2. Preprocessing
tokenizer = Tokenizer(num_words=100)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
data = pad_sequences(sequences, maxlen=5)

# 3. Model Architecture
model = Sequential([
    Embedding(input_dim=100, output_dim=8, input_length=5),
    LSTM(16),
    Dense(1, activation='sigmoid')
])

# 4. Compile and Train
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(data, labels, epochs=10, verbose=0)

# 5. Quick Test
test_txt = ["I love experience"]
test_seq = pad_sequences(tokenizer.texts_to_sequences(test_txt), maxlen=5)
prediction = model.predict(test_seq, verbose=0)

print(f"Prediction for '{test_txt[0]}': {'Positive' if prediction > 0.5 else 'Negative'} ({prediction[0][0]:.2f})")
```

9. Using spaCy, build a simple NLP pipeline that includes tokenization,
lemmatization, and entity recognition. Use the following paragraph as your dataset:

“Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the development of India’s atomic energy program. He was the founding director of the Tata Institute of Fundamental Research (TIFR) and was instrumental in establishing the Atomic Energy Commission of India.”

Write a Python program that processes this text using spaCy, then prints tokens, their lemmas, and any named entities found.

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of India’s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India."""

# 2. Process the text
doc = nlp(text)

# 3. Print Tokens and Lemmas (First 10 for brevity)
print(f"{'Token':<20} | {'Lemma':<20}")
print("-" * 45)
for token in list(doc)[:10]:
    print(f"{token.text:<20} | {token.lemma_:<20}")

# 4. Print Named Entities
print("\nNamed Entities:")
print("-" * 45)
for ent in doc.ents:
    print(f"{ent.text:<40} | {ent.label_}")
```
